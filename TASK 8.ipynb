```python
#import libraries
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
basket_Portugal = pd.read_csv('C:/Users/Admin/Desktop/Data Minning/Portugal_online_retail.csv')
```

```python
basket_Portugal
```

```python
#drop InvocieNO
basket_Portugal=basket_Portugal.drop('InvoiceNo',axis=1)
```

```python
# Finding the frequent itemsets
freq_items = apriori(basket_Portugal, min_support = 0.1, use_colnames = True)
```

```python
#Association rule mining
rules = association_rules(freq_items, metric= 'confidence', min_threshold=0.2)
```

```python
#selecting a new confidence level
new_confidence_level= 0.2
```

```python
#Filter rules based on new confidence level
important_rules = rules[rules['confidence']>= new_confidence_level]
```

```python
# Sort the rules by confidence and select the top three rules for each dataset
important_rules_sorted = important_rules.sort_values(by='confidence', ascending=False).groupby('antecedents').head(3)
```

```python
# Display the first three most important rules for each dataset
print(important_rules_sorted.head(3))
```
